```python
def get_locs(product):
    count = 0
    loc = []
    for i in range(len(df)):
        if df["product_group"][i] == product:
            count += 1
            loc.append(i)
    return loc

```

```python
def get_mean(product):
    loc = get_locs(product=product)
    total = 0
    for i in loc:
        total += df["unit_price"][i]
        mean = total/len(loc)
    return mean

```

```python
print(f"food average: {get_mean(product="Food")}")
print(f"beverage average: {get_mean(product='Beverages')}")
```

```python
def classify(df):
    classified = {
        "Baby_Boomers": 0,
        "Gen_X": 0,
        "Gen_Z": 0,
        "Older_Millenials": 0,
        "Younger_Millenials": 0
    }
    for i in range(len(df)):
        generation = df["customer_generation"][i]
        
        if generation == "Younger Millennials":
            classified["Younger_Millenials"] += 1
        elif generation == "Older Millennials":
            classified["Older_Millenials"] += 1
        elif generation == "Baby Boomers":
            classified["Baby_Boomers"] += 1
        elif generation == "Gen X":
            classified["Gen_X"] += 1
        elif generation == "Gen Z":
            classified["Gen_Z"] += 1 

    return classified


classified = classify(df=df)
print(classified)
```

```python
total = 0
for generation in classified:
    total += classified[generation]
gen_z = classified["Gen_Z"]

print((gen_z/total)*100)
```
